```python
import pandas as pd
import numpy as np
import random
import string
np.random.seed(56)
```

```python
def id():
    return ''.join(random.choice(string.ascii_letters) for i in range(5))
```

```python
def score():
    return np.random.randint(0,100)
```

```python
df = pd.DataFrame([[score() for i in range(4)]for i in range(10)],
                  index = [id() for i in range(10)],
                  columns = ['국어', '수학', '영어', '사회'])
```
